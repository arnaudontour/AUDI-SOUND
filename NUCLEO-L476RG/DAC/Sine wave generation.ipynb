## Sine wave form generation (fixed frequency)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def format_for_TreuStudio(y):
    END = ''
    l = len(y)-1
    print('{{{},'.format(y[0]) ,end='')
    for i in range(1,l):
        if (i % 8) == 0:
            END = '\n'
        else:
            END = ''
        print('{},'.format(y[i]) ,end=END)
    print('{}}};'.format(y[l]))
```

```python
Ns = 128
A = 128
x = np.linspace(0, Ns-1, Ns)
y = np.sin((2 * np.pi * x)/ Ns) * A + 2048
y = y.astype(np.uint16)
plt.plot(x,y)
```

```python
format_for_TreuStudio(y)
```
